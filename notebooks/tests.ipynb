```python
import numpy as np
embeddings = np.loadtxt("../embeddings/vectors.csv", delimiter=",")
query = np.loadtxt("../embeddings/queries/query.csv", delimiter=",")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(query.reshape(1, -1), embeddings)

similarity
```

```python
# index of the 10 most similar documents
similar_indices = similarity.argsort()[0][-10:][::-1]
similar_indices
```
